```python
allSL=[df_BMSL,df_BioSL,df_CALLSL,df_CALitSL,df_ChemSL,df_EBSL,df_ESSSL,df_EconSL,df_GeoSL,df_HisSL,df_MAASL,df_MAISL,df_MusicSL,df_PhiSL,df_PhysSL,df_VASL]
allHL=[df_BMHL, df_BioHL, df_MAAHL, df_CALLHL, df_CALitHL, df_ChemHL, df_EALitHL, df_EBHL, df_EconHL, df_GeoHL, df_HisHL, df_MAIHL, df_MusicHL, df_PhysHL, df_VAHL]
allCourse= allSL+allHL
nameSL=['BMSL','BioSL','CALLSL','CALitSL','ChemSL','EBSL','ESSSL','EconSL','GeoSL','HisSL','MAASL','MAISL','MusicSL','PhiSL','PhysSL','VASL']
nameHL=['BMHL', 'BioHL', 'MAAHL', 'CALLHL', 'CALitHL', 'ChemHL', 'EALitHL', 'EBHL', 'EconHL', 'GeoHL', 'HisHL', 'MAIHL', 'MusicHL', 'PhysHL', 'VAHL'] 
allNames=nameSL+nameHL

#Calculate the mean score of each course

for i in range(31):
    allCourse[i].columns=['score','id']
    print(allNames[i], allCourse[i]['score'].mean(axis=0))

courses= pd.concat([df_CALLHL, df_EBSL, df_MAASL,df_PhysHL, df_BMSL, df_GeoHL],axis=0)
print(courses['score'].mean(axis=0))
```

```python
#Calculate the mean score of each student
#create [id, mean score, course selction] list
for i in range(31):
    allCourse[i].columns=[allNames[i],'id']
df1=df_BMSL
for i in range(1,31):
    df2=allCourse[i]
    df1 = pd.merge(df1, df2, how="outer", on=["id"])
    
df_mean=df1[allNames].mean(1)

comb=''
courses=pd.DataFrame(data={'course selection': ['']})
for i in range(183):
    comb=''
    for course in allNames:
        score= df1.loc[i,course]
        if pd.notna(score):
            comb=comb+course+" "
    courses.loc[i]=comb 
            
idScore=pd.concat([df1['id'], df_mean, courses], axis=1)
idScore.columns=['id','mean score','course selection']
idScore=idScore.sort_values('mean score')
print(idScore)

idScore.to_csv('allstudentsscores.csv',index=False)

grouped=idScore.groupby("course selection")['mean score'].mean().reset_index(name='mean score of this combination')

grouped=grouped.sort_values('mean score of this combination')
print(grouped)

grouped.to_csv('mean score of all combination.csv',index=False)

numbers=idScore.groupby("course selection")['id'].count().reset_index(name='number of students')
numbers=numbers.sort_values('number of students')
print(numbers)
numbers.to_csv('numbers of students selecting each combination.csv',index=False)
```

```python
#load the [number of students, mean score] list fir HL subjects
df_means = pd.DataFrame(pd.read_csv('/kaggle/input/mean-and-count-of-hl/mean and number of HL subjects.txt')) 

# fit a simple linear regression
import statsmodels.api as sm
df_means.columns = ['mean','number']
mdl1 = sm.OLS(df_means['mean'],df_means['number'] )
mdl_fit = mdl1.fit()
print(mdl_fit.summary())

# correlation
# first check with a scatter plot
import matplotlib.pyplot as plt
plt.plot(df_means['number'], df_means['mean'], 'o', color='black');
plt.xlabel('number of students')
plt.ylabel('mean score')

# (optional) add the best fitted line to scatter plot
m, b = np.polyfit(df_means['number'], df_means['mean'], 1)
plt.plot(df_means['number'], m * df_means['number'] + b)
```

```python
#identify outliers in each HL class
for df in allHL:
    # calculate IQR for column Height
    Q1 = df['Scaled total'].quantile(0.25)
    Q3 = df['Scaled total'].quantile(0.75)
    IQR = Q3 - Q1
    # identify outliers
    threshold = 1.5
    outliers = df[(df['Scaled total'] < Q1 - threshold * IQR) | (df['Scaled total'] > Q3 + threshold * IQR)]
    print(outliers)
```
